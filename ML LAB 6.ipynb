```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, normalize
import scipy.cluster.hierarchy as shc
```

```python
X=pd.read_csv("/content/credit.csv")
#dropping the CUST_ID column from data
X=X.drop('CUST_ID',axis=1)
#handling missing values
X.fillna(method='ffill', inplace=True)
```

```python
#scaling the data so that all features become comparable
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
#Normalizing data so that the data apporx follows a gaussian distribution
X_normalized=normalize(X_scaled)
#convering the numpy array into a pandas dataframe
X_normalized=pd.DataFrame(X_normalized)
```

```python
#visualizing the data
plt.figure(figsize=(8,8))
plt.title("Visualizing the data")
Dendogram=shc.dendrogram((shc.linkage(X_normalized,method='average')))
```

At a particular point you can determine how many clusters will be formed by plotting a line and counting the number of crosssection intersections

```python
#PCA
ac2=AgglomerativeClustering(n_clusters=3)
pca=PCA(n_components=2)
X_principal=pca.fit_transform(X_normalized)
X_principal=pd.DataFrame(X_principal)
X_principal.columns=['P1','P2']
#Visualizing the clusters
plt.figure(figsize=(6,6))
plt.title("Visualizing the clustering")
plt.scatter(X_principal['P1'], X_principal['P2'], c=ac2.fit_predict(X_principal),cmap='rainbow')
```

```python
dataset=pd.read_csv("/content/aggl.csv")

X1=dataset.iloc[:,[3,4]].values
```

```python
dendogram1=shc.dendrogram(shc.linkage(X1,method='ward'))
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
model.fit(X1)
labels=model.labels_
labels
```

```python
dendogram2=shc.dendrogram(shc.linkage(X1,method='average'))
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='average')
model.fit(X1)
labels=model.labels_
labels
```

```python
dendogram3=shc.dendrogram(shc.linkage(X1,method='complete'))
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
model.fit(X1)
labels=model.labels_
labels
```
